* **units** : DataFrame | index 'unit_id' et colonnes 'waveform_PT_ratio', 'waveform_amplitude', 'amplitude_cutoff', 'cluster_id', 'cumulative_drift', 'd_prime', 'firing_rate', 'isi_violations', 'isolation_distance', 'L_ratio', 'local_index', 'max_drift', 'nn_hit_rate', 'nn_miss_rate', 'peak_channel_id', 'presence_ratio', 'waveform_recovery_slope', 'waveform_repolarization_slope', 'silhouette_score', 'snr', 'waveform_spread', 'waveform_velocity_above', 'waveform_velocity_below', 'waveform_duration', 'filtering', 'probe_channel_number', 'probe_horizontal_position', 'probe_id', 'probe_vertical_position', 'structure_acronym', 'ecephys_structure_id', 'ecephys_structure_acronym', 'anterior_posterior_ccf_coordinate', 'dorsal_ventral_ccf_coordinate', 'left_right_ccf_coordinate', 'probe_description', 'location', 'probe_sampling_rate', 'probe_lfp_sampling_rate', 'probe_has_lfp_data'

* **get_stimulus_table** : method | get_stimulus_table(stimulus_names: Any | None = None, include_detailed_parameters: bool = False, include_unused_parameters: bool = False) -> Any Get a subset of stimulus presentations by name, with irrelevant parameters filtered off
  * Retourne toutes les infos sur les présentations des stimulus (nom, temps, paramètres et condition_id)
* stimulus_conditions : DataFrame | index 'stimulus_condition_id' et colonnes 'size', 'color', 'mask', 'stimulus_name', 'opacity', 'units', 'temporal_frequency', 'contrast', 'y_position', 'orientation', 'phase', 'x_position', 'frame', 'spatial_frequency', 'color_triplet'

### Données brutes
* spike_amplitudes : dict | amplitude des spikes pour chaque unit (taille des arrays variable)
  * Contient les amplitudes des spikes de tous les units (attention, longueurs variables)
* spike_times : dict | temps des spikes pour chaque unit (taille des arrays variable)
  * Contient les temps des spikes de tous les units (attention, longueurs variables)

### Données groupées
* conditionwise_spike_statistics : method | conditionwise_spike_statistics(stimulus_presentation_ids: Any | None = None, unit_ids: Any | None = None, use_rates: bool = False) -> DataFrame Produce summary statistics for each distinct stimulus condition
  * Retourne des statistiques descriptives selon un unit et une présentation comme le spike_count, le nombre de présentations, le nombre moyen de spike count, etc.
* **presentationwise_spike_counts** : method | presentationwise_spike_counts(bin_edges: Any, stimulus_presentation_ids: Any, unit_ids: Any, binarize: bool = False, dtype: Any | None = None, large_bin_size_threshold: float = 0.001, time_domain_callback: Any | None = None) -> DataArray Build an array of spike counts surrounding stimulus onset per unit and stimulus frame.
  * Regroupe les spike counts en bins de temps (pour une présentation et un unit)
* **presentationwise_spike_times** : method | presentationwise_spike_times(stimulus_presentation_ids: Any | None = None, unit_ids: Any | None = None) -> DataFrame Produce a table associating spike times with units and stimulus presentations
  * Retourne pour une présentation les spike times de tous les units et le temps depuis le stimulus onset

## Réponses

```python
units = session.units
units.ecephys_structure_acronym.value_counts(dropna=False)
```

```python
VIS_units = units.query("ecephys_structure_acronym in ['VISp','VISl','VISal','VISrl','VISpm','VISam']")
VIS_units.head(5)
```

```python
counts = session.presentationwise_spike_counts(np.linspace(0,static_gratings.duration.max(),101),static_gratings.index,VIS_units.index)
counts
```

Les données sont de taille (presentation, n_bins, n_units).  
Les index sont stimulus_presentation_id, time_relative_to_stimulus_onset et unit_id. 

```python
plt.imshow(counts.data[0].T,aspect="auto")
plt.colorbar()
plt.show()
```
