# Exploration des sessions

```python
import warnings
warnings.filterwarnings("ignore")
from allensdk.brain_observatory.ecephys.ecephys_project_cache import EcephysProjectCache
```

```python
cache = EcephysProjectCache.from_warehouse(manifest="./data/manifest.json")
sessions = cache.get_session_table()
sessions.head(5)
```

Variables importantes : session_type, ecephys_structure_acronyms, full_genotype, age_in_days, sex

On sait qu'on ne regarde que les sessions du type brain_observatory_1.1	

```python
sessions_keep = sessions.query("session_type=='brain_observatory_1.1'").copy()
```

## Sessions par ecephys_structure

```python
sessions_keep.ecephys_structure_acronyms.value_counts(dropna=False)
```

Les régions les plus affectées par les static_gratings et les natural_scenes semblent être les VIS (https://observatory.brain-map.org/visualcoding/).  
Toutes les sessions brain_observatory_1.1 ciblent au moins une région VIS.  
La région la plus activée est VISp (toujours activée). Les autres importantes semblent être VISl, VISal, VISrl, VISpm et VISam. 

```python
sessions_keep[sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISp")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISl")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISal")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISrl")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISpm")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISam")]
```

```python
sessions_keep = sessions_keep[sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISp")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISl")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISal")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISrl")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISpm")*sessions_keep.ecephys_structure_acronyms.astype(str).str.contains("VISam")].copy()
sessions_keep["N_STRUCTURES"] = sessions_keep.ecephys_structure_acronyms.apply(lambda x: len(x))
```
