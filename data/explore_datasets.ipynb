```python
with np.load("natural_scenes_dataset_500_binned.npz") as data:
    X = data["X"]
    y = data["y"]

print(X.shape)
print(y.shape)
```

```python
first_image = X[0]
first_neuronal_activation = y[0]

print(first_image.shape)
print(first_neuronal_activation.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.imshow(first_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# shape should be (152, 18)
data = first_neuronal_activation

plt.figure(figsize=(12, 10))
plt.imshow(data, aspect="auto", cmap="viridis")

plt.colorbar(label="Activation")
plt.xlabel("Timestamp")
plt.ylabel("Neuron")
plt.title("Neuronal Activity (152 neurons x 18 timepoints)")
plt.show()
```

```python
def show_neuronal_grid(y, n_rows=3, n_cols=3):
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 12), constrained_layout=True)

    axes = axes.flatten()

    vmin, vmax = np.min(y), np.max(y)

    for i in range(n_rows * n_cols):
        if i >= len(y):
            axes[i].axis("off")
            continue

        data = y[i+5]

        im = axes[i].imshow(data, aspect="auto", cmap="viridis")
        axes[i].axis("off")
        axes[i].set_title(f"Neuronal activity {i + 5}", fontsize=8)

    fig.colorbar(im, ax=axes.ravel().tolist(), label="Activation")
    #plt.title("Neuronal Activity (152 neurons x 18 timepoints)")
    #plt.tight_layout()
    plt.show()
```

```python
show_neuronal_grid(y)
```

Let's look at more images

```python
def show_image_grid(X, n_rows=3, n_cols=4):
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 8))

    axes = axes.flatten()

    for i in range(n_rows * n_cols):
        if i >= len(X):
            axes[i].axis("off")
            continue

        axes[i].imshow(X[i], cmap="gray")
        axes[i].axis("off")
        axes[i].set_title(f"img {i}", fontsize=8)

    plt.tight_layout()
    plt.show()
```

```python
show_image_grid(X, n_rows=5, n_cols=5)
```
