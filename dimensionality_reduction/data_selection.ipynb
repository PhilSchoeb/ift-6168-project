## Première expérience

Session : 750332458  
On ne garde que les stimulus avec barres horizontales ou verticales.  
On ne garde que les 100 units les plus expressifs (ceux qui différencient le mieux les 2 types de stimulus). 

````python
session = 750332458
````

````python
dataset = StaticGratingsDataset(session,cache_dir="../data/ecephys/session_data")
dataset
````

````python
h_v_bars = dataset.get_presentation_ids(orientation=[0,90])
X, y = dataset.get_data(presentation_ids=h_v_bars,stimulus_type="params")
X.shape, y.shape
````

````python
stimulus_table = dataset.stimulus_table.loc[h_v_bars].copy()
stimulus_table["class"] = np.where(stimulus_table.orientation==0,0,1)
stimulus_table.reset_index(inplace=True)
stimulus_table
````

````python
results = np.empty((2,100,295))
for c in range(2):
    idx = stimulus_table[stimulus_table["class"]==c].index
    array = y[idx,:,:].sum(axis=0,keepdims=True)
    results[c,:,:] = array/len(idx)
final = results.var(axis=0).mean(axis=0)
final.shape
````

````python
variances = pd.DataFrame(final,index=dataset.units.index,columns=["Mean variance"]).sort_values("Mean variance",ascending=False)
variances
````

````python
most_expressive = variances.index[:100]
most_expressive
````

````python
X_final, y_final = dataset.get_data(presentation_ids=h_v_bars,unit_ids=most_expressive,stimulus_type="params")
X_final.shape, y_final.shape
````

````python
plot_data_samples(X_final,y_final,4)
````

````python
os.makedirs(f"./session_data/session_{session}",exist_ok=True)
np.savez(f"./session_data/session_{session}/static_gratings.npz",X=X_final,y=y_final,presentation_ids=np.array(h_v_bars),unit_ids=np.array(most_expressive),timestamps=dataset.timestamps)
````
